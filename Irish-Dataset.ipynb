```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
filename = '/content/drive/My Drive/Colab_Test/Dataset/Clustering/iris.csv'

dataset = pd.read_csv(filename)

dataset = dataset.drop(['Id', 'Species'], axis=1)
```

```
dataset.head()
```

```
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=10)
    kmeans.fit(dataset)
    wcss.append(kmeans.inertia_) # Within cluster sum of squares

print(wcss)
```

```
plt.plot(range(1, 11), wcss)
plt.title('Elbow method')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
```

```
kmeans = KMeans(n_clusters=3, random_state=10)
model = kmeans.fit(dataset)
```

```
model.labels_
```

```
model.cluster_centers_
```

```
y_kmeans = kmeans.predict(dataset)
```

```
# SepalLengthCm dan	SepalWidthCm

plt.scatter(dataset.values[y_kmeans == 0, 0], dataset.values[y_kmeans == 0, 1], s=60, c='red', label ='Cluster 1')
plt.scatter(dataset.values[y_kmeans == 1, 0], dataset.values[y_kmeans == 1, 1], s=60, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[y_kmeans == 2, 0], dataset.values[y_kmeans == 2, 1], s=60, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, marker='*', label = 'Centeroids', c='yellow')
plt.legend(loc='best')
```

#Dari data diatas, korelasi antara SepalLengthCm dengan SepalWidthCm sangat baik, dikarenakan kita melihat pengelompokan dengan jelas pada gambar.

```
# SepalLengthCm dan PetalLengthCm

plt.scatter(dataset.values[y_kmeans == 0, 0], dataset.values[y_kmeans == 0, 2], s=60, c='red', label ='Cluster 1')
plt.scatter(dataset.values[y_kmeans == 1, 0], dataset.values[y_kmeans == 1, 2], s=60, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[y_kmeans == 2, 0], dataset.values[y_kmeans == 2, 2], s=60, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 2], s=300, marker='*', label = 'Centeroids', c='yellow')
plt.legend(loc='best')
```

#Pada gambar diatas, korelasi antara SepalLengthCm dengan PetalLengthCm sangatlah baik, dikarenakan pengelompokan yang jelas dapat dilihat langsung pada gambar.

```
# SepalWidthCm dan PetalWidthCm

plt.scatter(dataset.values[y_kmeans == 0, 1], dataset.values[y_kmeans == 0, 3], s=60, c='red', label ='Cluster 1')
plt.scatter(dataset.values[y_kmeans == 1, 1], dataset.values[y_kmeans == 1, 3], s=60, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[y_kmeans == 2, 1], dataset.values[y_kmeans == 2, 3], s=60, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 1], model.cluster_centers_[:, 3], s=300, marker='*', label = 'Centeroids', c='yellow')
plt.legend(loc='best')
```
